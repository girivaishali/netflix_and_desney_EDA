# Importing the Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

theme = ["#d2fbd4","#a5dbc2","#7bbcb0","#559c9e","#3a7c89","#235d72","#123f5a"]
```

# Load and Clean the Data

```python
Disney=pd.read_csv('disney_plus_titles.csv')
```

```python
Disney.head()
```

```python
Disney = Disney.drop(['show_id','cast','description'], axis = 1)
```

```python
Disney.head()
```

```python
Netflix=pd.read_csv('titles.csv')
```

```python
Netflix.loc[Netflix['type'] == 'SHOW','type']= 'TV Show'
Netflix.loc[Netflix['type'] == 'MOVIE','type'] = 'Movie'
Netflix['genres'] = Netflix['genres'].str.replace('\'', '')
Netflix['genres'] = Netflix['genres'].str.replace('\'', '')
Netflix['production_countries'] = Netflix['production_countries'].str.replace('\'', '')
Netflix['production_countries'] = Netflix['production_countries'].str.replace('\'', '')

Netflix['genres'] = Netflix['genres'].str.replace(']', '')
Netflix['genres'] = Netflix['genres'].str.replace('[', '')
Netflix['production_countries'] = Netflix['production_countries'].str.replace('[', '')
Netflix['production_countries'] = Netflix['production_countries'].str.replace(']', '')
Netflix.head()
```

```python
Netflix = Netflix.drop(['id','description'], axis = 1)
```

```python
Netflix.head()
```

# Analyzing the data

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 9))
fig.suptitle(' Content Comparision ', size = 26, color = theme[5], weight='bold')
axs = [ax1, ax2]

#figure1
fig1 = sns.countplot(x="type",data=Netflix,ax=ax1,palette="crest", order=['TV Show', 'Movie'])
fig1.set_ylabel("Count",fontsize=10, weight='bold')
fig1.set_xlabel("Type",fontsize=10,weight='bold')
ax1.set_title(f'Netflix Total: {Netflix.type.value_counts().sum()}', size = 14, color = theme[3],weight='bold', pad=20)


#figure2

fig2=sns.countplot(x="type", data=Disney,ax=ax2,palette="crest", order=['TV Show', 'Movie'])
fig2.set_ylabel("Count",fontsize=10, weight='bold')
fig2.set_xlabel("Type",fontsize=10,weight='bold')
ax2.set_title(f'Disney Total: {Disney.type.value_counts().sum()}', size = 14,color = theme[3], weight='bold', pad=20)

plt.subplots_adjust(left=None, bottom=None, right=None, top=0.8, wspace=0.4, hspace=None);
plt.show()
```

```python
fig_dims=(15,8)

fig, ax = plt.subplots(figsize=fig_dims)
rat1 = sns.histplot(x= Netflix['release_year'],data=Netflix ,hue = 'type',palette = 'crest',bins=list(range(1930,2020,10)))
rat1.legend(['Movie','TV Show'],loc='upper right')
rat1.set_ylabel("Count",fontsize=15, weight='bold')
rat1.set_xlabel("Release Year",fontsize=15,weight='bold')
rat1.figure.suptitle('Count of Movies and TV Shows Content by Release Year on NETFLIX',y=1, size = 20, color = theme[5], weight='bold');
```

```python
fig, ax = plt.subplots(figsize=fig_dims)
rat1 = sns.histplot(x= Disney['release_year'],data=Disney ,hue = 'type',palette = 'crest',bins=list(range(1930,2020,10)))
rat1.legend(['Movie','TV Show'],loc='upper right')
rat1.set_ylabel("Count",fontsize=15, weight='bold')
rat1.set_xlabel("Release Year",fontsize=15,weight='bold')
rat1.figure.suptitle('Count of Movies and TV Shows Content by Release Year on DISNEY',y=1, size = 20, color = theme[5], weight='bold');
```

```python
country_type = Netflix[Netflix['production_countries'] != "unknown"].groupby(by = ['production_countries','type'])['production_countries'].count().to_frame()
country_type.columns = ['count']
country_type.reset_index()
country_type = country_type.sort_values(by=['count','production_countries'] ,  ascending = False).head(15).reset_index()


fig, ax = plt.subplots(figsize=fig_dims)
rat1 = sns.barplot(data=  country_type , x = 'production_countries' , y = 'count' , hue = 'type', palette = 'crest');
rat1.legend(['Movie','TV Show'],loc='upper right')
rat1.set_ylabel("Count",fontsize=15, weight='bold')
rat1.set_xlabel("Production Countries",fontsize=15,weight='bold')
rat1.figure.suptitle('Count of Movies and TV Shows Content by Countries on NETFLIX',y=1, size = 20, color = theme[5], weight='bold');
```

```python
country_type = Disney[Disney['country'] != "unknown"].groupby(by = ['country','type'])['country'].count().to_frame()
country_type.columns = ['count']
country_type.reset_index()
country_type = country_type.sort_values(by=['count','country'] ,  ascending = False).head(10).reset_index()


fig, ax = plt.subplots(figsize=fig_dims)
rat1 = sns.barplot(data=  country_type , x = 'country' , y = 'count' , hue = 'type', palette = 'crest');
rat1.legend(['Movie','TV Show'],loc='upper right')
rat1.set_ylabel("Count",fontsize=15, weight='bold')
rat1.set_xlabel("Production Countries",fontsize=15,weight='bold')
rat1.figure.suptitle('Count of Movies and TV Shows Content by Countries on DISNEY',y=1, size = 20, color = theme[5], weight='bold');
plt.xticks(rotation = 90);
```

```python
fig_dims = (15, 8)

fig, ax = plt.subplots(figsize=fig_dims)
rat1 = sns.countplot(x= Netflix['age_certification'],data=Netflix,hue = 'type',palette = 'crest')
rat1.legend(['Movie','TV Show'],loc='upper right')
rat1.set_ylabel("Count",fontsize=15, weight='bold')
rat1.set_xlabel("Rating Type",fontsize=15,weight='bold')
rat1.figure.suptitle('Count of Video content rating on NETFLIX',y=1, size = 20, color = theme[3], weight='bold');
```

```python
fig, ax = plt.subplots(figsize=fig_dims)
rat2 = sns.countplot(x= Disney['rating'],data=Disney,hue = 'type',palette='crest')
rat2.set_ylabel("Count",fontsize=15, weight='bold')
rat2.set_xlabel("Rating Type",fontsize=15,weight='bold')
rat2.figure.suptitle('Count of Video content rating on Disney',y=1, size = 20, color = theme[3], weight='bold');
```

```python
Netflix_genere_list = Netflix.genres[Netflix.genres.notnull()].tolist()
Netflix_genere_string_list = ",".join(Netflix_genere_list).replace(" ","")
Netflix_genere = Netflix_genere_string_list.split(",")


n_unique,n_counts = np.unique(Netflix_genere, return_counts=True)

Netflix_unique  = pd.DataFrame()
Netflix_unique['genere'] = n_unique
Netflix_unique['count'] = n_counts
Netflix_unique = Netflix_unique.sort_values(by = 'count',ascending = False)[:10]



fig, ax = plt.subplots(figsize=fig_dims)
rat1 = sns.barplot(data=  Netflix_unique , x = 'genere' , y = 'count', palette = 'crest');
rat1.set_ylabel("Count",fontsize=15, weight='bold')
rat1.set_xlabel("Genere",fontsize=15,weight='bold')
rat1.figure.suptitle('Top 10 Generes on NETFLIX',y=1, size = 20, color = theme[5], weight='bold');
```

```python
Disney_genere_list = Disney.listed_in[Disney.listed_in.notnull()].tolist()
Disney_genere_string_list = ",".join(Disney_genere_list).replace(" ","")
Disney_genere = Disney_genere_string_list.split(",")

d_unique,d_count = np.unique(Disney_genere, return_counts=True)
Disney_unique  = pd.DataFrame()
Disney_unique['genere'] = d_unique
Disney_unique['count'] = d_count
Disney_unique = Disney_unique.sort_values(by = 'count',ascending = False)[:10]



fig, ax = plt.subplots(figsize=fig_dims)
rat1 = sns.barplot(data=  Disney_unique , x = 'genere' , y = 'count', palette = 'crest');
rat1.set_ylabel("Count",fontsize=15, weight='bold')
rat1.set_xlabel("Genere",fontsize=15,weight='bold')
rat1.figure.suptitle('Top 10 Generes on Disney',y=1, size = 20, color = theme[5], weight='bold');
```

*Netflix has more content and the content is more available in non US countries.

*If we consider the content which was released after 2000 as "new", Netflix has more uptodate content compared to the Disney Plus.

*Both platforms have similar ratios of Movies to TV Shows.
